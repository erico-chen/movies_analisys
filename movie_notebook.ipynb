# Movie Analisys

The database used in the analisys was pulled out from [kaggle](https://www.kaggle.com/). It contains movies data from 2000 to 2024. but not all the movies released in this time, only the top 200 (in box office) by each year. The database has 5000 rows and 13 columns, those are:
1. Rank: Movie's rank
2. Release Group: Movie's name
3. \$Worldwide: Worldwide box office
4. \$Domestic: Domestic box office
5. Domestic %: The percentage of the domestic market in worldwide box office 
6. \$Foreign: Foreign box office
7. Foreign %: The percentage of the foreign market in worldwide box office 
8. Year: Year the movie was released
9. Genres: Movie's genre
10. Rating: The average rating from TMDB
11. Vote_Count: Number of votes the movie had
12. Original_Language: Movie's original language
13. Production_Countries: Country (ies) where the movie was filmed
